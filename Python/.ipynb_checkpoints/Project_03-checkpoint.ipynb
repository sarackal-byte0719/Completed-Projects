##  Part 3: Working with Categorical Variables

In Part 3 section below, I will be creating lists to specify the order for each of the three categorical variables of cut, color and clarity. 

```python

clarity_levels = ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']

cut_levels = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']

color_levels = ['J', 'I', 'H', 'G', 'F', 'E', 'D']

```

In the cell below, I will use these lists to communicate to Pandas the correct order for the levels of the three categorical variables of clarity, cut and color.

```python
diamonds.cut = pd.Categorical(diamonds.cut, cut_levels, ordered = True)

diamonds.color = pd.Categorical(diamonds.color, color_levels, ordered = True)

diamonds.clarity = pd.Categorical(diamonds.clarity, clarity_levels, ordered = True)

```

In the cell below, I will create lists of named colors to serve as palettes to be used for visualizations that will be shown later in this notebook.

```python

clarity_pal = ['royalblue','g', 'r', 'c', 'm','darkorange','gray', 'gold']

cut_pal = ['maroon', 'orange', 'green', 'blue', 'purple']

color_pal = ['salmon', 'cornflowerblue', 'sandybrown', 'yellow', 'plum', 'teal', 'lawngreen']


```

## Part 5: Scatterplots of Price Against Carat

In part 5 section below, I will explore the relationship between the price and carat attributes by generating scatterplots.

```python
plt.figure(figsize=[8,6])
plt.scatter(x=diamonds['carat'], y=diamonds['price'], s = 20, alpha=0.4, color= 'blue') 
plt.xlabel('Carat Size')
plt.ylabel('Price (in $)')
plt.title ('Relationship between Diamond Price and Carat Size')
plt.show()


```

In the cell below, I will reproduce the plot from above, but I will also include color of the points based on their clarity.

```python
clarity_levels = ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']
clarity_pal = ['royalblue','g', 'r', 'c', 'm','darkorange','gray', 'gold']

plt.figure(figsize = [8,6])
for i, c in enumerate(clarity_levels):
    plt.scatter(diamonds.loc[diamonds.clarity == c,'carat'], 
                diamonds.loc[diamonds.clarity == c,'price'],
                s=20, alpha=0.4, label=c.title(),
                color=clarity_pal[i])

plt.xlabel('Carat Size')
plt.ylabel('Price (in $)')
plt.title ('Relationship between Diamond Price and Carat Size')
plt.legend()
plt.show()
    
 
    
```

In the cell below, I will separate the points for each of the eight clarity levels into its own subplot

```python
clarity_levels = ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']
clarity_pal = ['royalblue','g', 'r', 'c', 'm','darkorange','gray', 'gold']

plt.figure(figsize = [12,6])
for i, c in enumerate(clarity_levels):
    plt.subplot(2,4,i+1)
    plt.scatter(diamonds.loc[diamonds.clarity == c,'carat'], 
                diamonds.loc[diamonds.clarity == c,'price'],
                s=20, alpha=0.4, label=c.title(),
                color=clarity_pal[i])
    plt.xlim([0,4])
    plt.ylim([0,20000])
    plt.xlabel('Carat Size')
    plt.ylabel('Price (in $)')
    plt.title ('Clarity Level:'+ clarity_levels[i])

plt.tight_layout()
plt.show()
    
```

## Part 6: Applying Logarithmic Transformations

In Part 6 section below, I will apply the logarithmic transformations to the price and carat columns, create 2 new columns from these calculations, and will explore the distribution of these transformed variables.

```python
diamonds.loc[:,'ln_carat'] = np.log(diamonds.carat) 
diamonds.loc[:,'ln_price'] = np.log(diamonds.price)


```

In the cell below I will use histograms to explore the distribution of the diamond prices, and the log of the diamond prices.

```python
plt.figure(figsize=[12,4])

plt.subplot(1, 2, 1)
plt.hist(diamonds.loc[:,'price'], edgecolor='k', color = ['cyan'], 
         bins=20)
plt.title('Histogram of Prices')
plt.xlabel('Price')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
plt.hist(diamonds.loc[:,'ln_price'], edgecolor='k', color = ['deeppink'], 
         bins=20)
plt.title('Histogram of Log-Prices')
plt.xlabel('Natural Log of Price')
plt.ylabel('Count')

plt.show()




```

In the cell below, I will create histograms to explore the distribution of the carat sizes, and the log of the carat sizes.

```python
plt.figure(figsize=[12,4])

plt.subplot(1, 2, 1)
plt.hist(diamonds.loc[:,'carat'], edgecolor='k', color = ['tomato'], 
         bins=20)
plt.title('Histogram of Carat Sizes')
plt.xlabel('Carat Size')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
plt.hist(diamonds.loc[:,'ln_carat'], edgecolor='k', color = ['lime'], 
         bins=20)
plt.title('Histogram of Log-Carat Sizes')
plt.xlabel('Natural Log of Carat Size')
plt.ylabel('Count')

plt.show()

```

## Part 7: Scatterplot of Transformed Variables

In Part 7 section below, I will recreate the second scatterplot from Part 5 using the transformed variables( ln_price and ln_carat) instead of the original values.

In the cell below, I will create a scatterplot of ln_price against ln_carat, with points colored according to their clarity.

```python

clarity_levels = ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']
clarity_pal = ['royalblue','g', 'r', 'c', 'm','darkorange','gray', 'gold']

plt.figure(figsize = [8,6])
for i, c in enumerate(clarity_levels):
    plt.scatter(diamonds.loc[diamonds.clarity == c,'ln_carat'], 
                diamonds.loc[diamonds.clarity == c,'ln_price'],
                s=20, alpha=0.4, label=c.title(),
                color=clarity_pal[i])

plt.xlabel('Natural Log of Carat Size')
plt.ylabel('Natural Log of Price')
plt.title ('Relationship between Log-Price and Log-Carat Size')
plt.legend()
plt.show()



```

## Part 8: Grouping by Cut

In Part 8 section below, I will be calculating the mean price and carat size for each level of the variable cut.

In the cell below, I will be creating a grouped DataFrame called gb_cut to display the mean price and carat size for each level of cut. 

```python
gb_cut = diamonds.loc[:, ['cut', 'price', 'carat']].\
                  groupby(['cut']).mean()
gb_cut
```

In the cell below, I will create the bar charts to graphically display the information from the gb_cut dataFrame.

```python
cut_pal = ['maroon', 'orange', 'green', 'blue', 'purple']

plt.figure(figsize=[12,4])

plt.subplot(1, 2, 1)
plt.bar(x= gb_cut.index, height=gb_cut['price'], color = cut_pal, edgecolor='black')
plt.xlabel('Cut')
plt.ylabel('Mean Price')
plt.title('Mean Price by Cut')

plt.subplot(1, 2, 2)
plt.bar(x= gb_cut.index, height=gb_cut['carat'], color = cut_pal, edgecolor='black')
plt.xlabel('Cut')
plt.ylabel('Mean Carat Size')
plt.title('Mean Carat Size by Cut')

plt.show()



```
